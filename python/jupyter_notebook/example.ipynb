## A simple numerical integration example. 

Again we choose to demonstrate the use by a simple numerical integration estimate of $\pi$, although here we will use a different method via the integral equation
\begin{equation}
\pi = 4\int_0^1 \frac{1}{1+x^2} \:\mathrm{d}x.
% In Jupyter notebooks the LaTeX is generated using mathjax, so we can only use a limited range of functionality. 
\end{equation}
Hence to estimate this we use the simple numerical scheme:
\begin{equation}
\pi = 4\int_0^1 \frac{1}{1+x^2} \:\mathrm{d}x \approx 4\sum_{i=0}^{N-1} \frac{1}{1 + \left(\frac{i}{N}\right)^2} \frac{1}{N}
\end{equation}
for some large integer $N$.

```python
import numpy as np
import matplotlib.pylab as plt

M = 20 # The number of samples we generate (log-2 scale)
x = np.logspace(1, M, num=M,base=2)  # Number of samples
y = np.array([np.fabs(np.pi - 4.0*np.sum(1.0/(1.0 + \
    np.linspace(0,1,num=int(n+1))**2.0))/n)/(np.pi) for n in x]) \
    # Relative error.

plt.plot(x,y)
plt.title(r'Error in $\pi$-estimate from integral estimate')
plt.ylabel('Relative error')
plt.xlabel(r'Degree of discretisation $N$')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

We can see from this figure that the error decreases by $O(N^{-1})$. This is better than Monte Carlo, but there are some much better schemes for numerical integration of this kind.
